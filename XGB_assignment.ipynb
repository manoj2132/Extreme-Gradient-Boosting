```python
from sklearn.impute import SimpleImputer
df['Age']=df['Age'].fillna(df['Age'].mean)
df['Fare']=df['Fare'].fillna(df['Fare'].mean)
```

```python
df['Sex']=df['Sex'].map({'male':0,'female':1})
df['Sex']
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
x=df.drop(['Embarked','Cabin','Ticket','Name','PassengerId','Sex'],axis=1)
x
```

```python
y=df['Sex']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
print(x_train.dtypes)
```

```python
x_train['Age'] = pd.to_numeric(x_train['Age'], errors='coerce')
x_train['Fare'] = pd.to_numeric(x_train['Fare'], errors='coerce')
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
x_train['Age'] = imputer.fit_transform(x_train[['Age']])
x_train['Fare'] = imputer.fit_transform(x_train[['Fare']])
```

```python
x_train.shape,x_test.shape
```

```python
from xgboost import XGBClassifier
model =XGBClassifier(enable_categorical=True)
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
predict_train=model.predict(x_train)
```

```python
predict_train
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_train,predict_train)
```

```python
from sklearn.metrics import classification_report
print(classification_report(predict_train,y_train))
```

```python
from sklearn.metrics import accuracy_score
accuracy_train=accuracy_score(predict_train,y_train)
accuracy_train
```
